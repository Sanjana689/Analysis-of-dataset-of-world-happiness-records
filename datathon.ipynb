**Data visualisation**

```python
correlation = data0.corr()
```

**Heatmap visualisation**

1) It is a visualizing technique
that shows magnitude of a phenomenon as color in two dimensions. 

2) The variation in color is by hue (density) giving obvious visual cues to the reader.

i) Highest density value i.e. Economy(GDP per capita) vs Health (Life Expectency) has the lightest color

ii) The Least density value i.e. Happiness Rank vs Happiness Score has the darkest color.

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

**Scatter plot**

This plot compares 2 attributes at a time of the columns of the dataset.
It produces a matrix of relationshops between each variable in the data for an instant examination of a data.

Using the above visualizations, we have further done 
our analysis using Happiness Score VS Health(Life expectancy) and Happiness Score Vs Economy(GDP per capita)


*   We have taken these 2 attributes(i.e. Health and Economy) to compare which of the 2 leads to more happiness i.e. health or wealth.


*  We are also comparing trust in government with happiness Score to see if they have a relationship.






**Scatter plot of Heath vs Happiness score with respect to all the Countries** 

```python
sns.relplot(x='Happiness Score', y = 'Health (Life Expectancy)' , hue='Country', data=data0 )
```

**Scatter plot of Economy vs Happiness score with respect to Country**

```python
sns.relplot(x='Happiness Score', y = 'Economy (GDP per Capita)' , hue='Country', data=data0 )
```

**Scatter plot of trust vs Happiness score with respect to Country**

```python
sns.relplot(x='Happiness Score', y = 'Trust (Government Corruption)' , hue='Country', data=data0 )
```

**region wise bar graphs:-**

```python
sns.barplot(x=data0['Happiness Score'],y=data0['Economy (GDP per Capita)'])
```

```python
sns.barplot(x=data0['Happiness Score'],y=data0['Health (Life Expectancy)'])
```

```python
sns.barplot(x=data0['Happiness Score'],y=data0['Trust (Government Corruption)'])
```

**CONCLUSIONS**

From the above visualizations we have concluded that Economy (GDP per Capita) vs happiness Score has more linearity than Health(Life Expectency) Vs happiness Score. 

Also we can conclude that, in regions that have good Economy(GDP) and health, the happiness score is more. here, Western Europe has the highest happiness score.
Similarly, due to  low GDP and Health, Sub Saharan Africa has the least Happiness score.

Happiness of the people has a varying relationship with trust(government corruption).
Most of the countries have less trust(governmnet corruption).

## **muliple regression model-2016**

```python
data3.isnull().sum()
```

```python
dataset=data3.drop(['Country','Region'],axis=1)
dataset.head()
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
from sklearn.model_selection import train_test_split
```

```python
dependent_variable='Happiness Score'
```

```python
independent_variables=dataset.columns.tolist()
```

```python
independent_variables.remove(dependent_variable)
```

```python
independent_variables
```

```python
X=dataset[independent_variables].values
y=dataset[dependent_variable].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train[0:10]
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import math
math.sqrt(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
r1=r2_score(y_test,y_pred)
r1
```
